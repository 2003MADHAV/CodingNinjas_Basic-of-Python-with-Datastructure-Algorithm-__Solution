```python
# Problem ID 91, removeConsecutiveDuplicates
def removeConsecutiveDuplicates(string):

    # Base case: if the string is empty or has only one character, return the string
    if len(string) <= 1:
        return string

    # If the first character is the same as the second one, skip the first character
    if string[0] == string[1]:
        return removeConsecutiveDuplicates(string[1:])
    else:
        # If the first character is different from the second one, keep it and check the rest of the string
        return string[0] + removeConsecutiveDuplicates(string[1:])

string = input().strip()
print(removeConsecutiveDuplicates(string))
```

Merg Sort

```python
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = arr[l:l + n1]
    R = arr[m + 1:m + 1 + n2]

    i = j = 0
    k = l

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:
        m = (l + r) // 2

        mergeSort(arr, l, m)
        mergeSort(arr, m + 1, r)
        merge(arr, l, m, r)
```

Quick Sort

```python
from typing import List

def partition(arr, start, end):
    pivot = arr[end]
    i = start - 1

    for j in range(start, end):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[end] = arr[end], arr[i + 1]
    return i + 1

def quickSort(arr: List[int], start: int, end: int):
    if start < end:
        pivot = partition(arr, start, end)

        quickSort(arr, start, pivot - 1)
        quickSort(arr, pivot + 1, end)
```

```python
from math import *
from collections import *
from sys import *
from os import *

## Read input as specified in the question.
## Print output as specified in the question.
def countWays(n):
    # Create an array to store the number of ways for each step
    ways = [0] * (n + 1)
    ways[0] = 1  # There is one way to reach step 0 (base case)
    ways[1] = 1  # There is one way to reach step 1
    ways[2] = 2  # There are two ways to reach step 2

    # Calculate the number of ways for steps 3 to n using dynamic programming
    for i in range(3, n + 1):
        ways[i] = ways[i - 1] + ways[i - 2] + ways[i - 3]

    return ways[n]

# Read input integer N
N = int(input())

# Calculate the number of possible ways using the function
result = countWays(N)

# Print the number of possible ways
print(result)
```

```python
from sys import stdin

def pairSum(arr, n, num):
    frequency = {}
    count = 0

    for element in arr:
        complement = num - element
        if complement in frequency:
            count += frequency[complement]

        if element in frequency:
            frequency[element] += 1
        else:
            frequency[element] = 1

    return count


#taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n

#main
t = int(stdin.readline().strip())

while t > 0 :
    arr, n = takeInput()
    num = int(stdin.readline().strip())
    print(pairSum(arr, n, num))

    t -= 1
```

Triplet Sum

```python
def tripletSum(arr, n, num):
    arr.sort()
    count = 0

    for i in range(n - 2):
        left = i + 1
        right = n - 1

        while left < right:
            curr_sum = arr[i] + arr[left] + arr[right]

            if curr_sum == num:
                if arr[left] == arr[right]:
                    element_count = right - left + 1
                    count += (element_count * (element_count - 1)) // 2
                    break

                left_count = 1
                right_count = 1

                while left + 1 < right and arr[left] == arr[left + 1]:
                    left += 1
                    left_count += 1

                while right - 1 > left and arr[right] == arr[right - 1]:
                    right -= 1
                    right_count += 1

                count += left_count * right_count
                left += 1
                right -= 1

            elif curr_sum < num:
                left += 1
            else:
                right -= 1

    return count


# taking input using fast I/O method
def takeInput():
    n = int(input())
    if n == 0:
        return list(), 0

    arr = list(map(int, input().strip().split()))
    return arr, n


# main
t = int(input())

while t > 0:
    arr, n = takeInput()
    num = int(input())
    print(tripletSum(arr, n, num))

    t -= 1
```

Append last N to first

```python

from sys import stdin

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def appendLastNToFirst(head, n):
    if head is None or n == 0:
        return head

    # Count the length of the linked list
    length = 0
    temp = head
    while temp is not None:
        length += 1
        temp = temp.next

    # Edge case: If n is greater or equal to the length of the list
    if n >= length:
        return head

    k = length - n
    prev = None
    current = head

    # Traverse the list to find the 'k-th' node
    while k > 0:
        prev = current
        current = current.next
        k -= 1

    # Change the links to move the last 'N' nodes to the front
    temp = current
    while temp.next is not None:
        temp = temp.next
    temp.next = head
    head = current
    prev.next = None

    return head



#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    n = int(stdin.readline().rstrip())

    head = appendLastNToFirst(head, n)
    printLinkedList(head)

    t -= 1
```

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def skipMdeleteN(head, M, N):
    if not head or M == 0:
        return None

    current = head
    prev = None

    while current:
        m_count = 0
        n_count = 0

        # Skip M nodes
        while current and m_count < M:
            prev = current
            current = current.next
            m_count += 1

        # Delete N nodes
        while current and n_count < N:
            current = current.next
            n_count += 1

        # Adjust connections after deleting N nodes
        if prev:
            prev.next = current

    return head





#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    m_n = stdin.readline().strip().split(" ")

    m = int(m_n[0])
    n = int(m_n[1])

    newHead = skipMdeleteN(head, m, n)
    printLinkedList(newHead)

    t -= 1
```

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def swapNodes(head, i, j):
    if i == j:
        return head

    # Track the current and previous nodes for i and j
    prev_i = None
    current_i = head
    while i > 0 and current_i:
        prev_i = current_i
        current_i = current_i.next
        i -= 1

    prev_j = None
    current_j = head
    while j > 0 and current_j:
        prev_j = current_j
        current_j = current_j.next
        j -= 1

    # Handle edge cases where i or j is out of range or equal
    if not current_i or not current_j:
        return head

    # Update connections to swap the nodes
    if prev_i:
        prev_i.next = current_j
    else:
        head = current_j

    if prev_j:
        prev_j.next = current_i
    else:
        head = current_i

    temp = current_i.next
    current_i.next = current_j.next
    current_j.next = temp

    return head


#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    i_j = stdin.readline().strip().split(" ")

    i = int(i_j[0])
    j = int(i_j[1])

    newHead = swapNodes(head, i, j)
    printLinkedList(newHead)

    t -= 1
```

```python
def checkRedundantBrackets(expression):
    stack = []

    # Traversing the expression
    for char in expression:
        if char == ')':
            top = stack.pop()  # Pop the top element of the stack

            # Check for a redundant expression inside brackets
            redundant = True
            while top != '(':
                if top in ['+', '-', '*', '/']:
                    redundant = False  # No redundant brackets, operators found
                top = stack.pop()

            if redundant:  # If a redundant bracket is found
                return True

        else:
            stack.append(char)  # Push non-closing brackets and operands to the stack

    return False

# Main
expression = input().strip()

if checkRedundantBrackets(expression):
    print("true")
else:
    print("false")
```

```python

from sys import stdin

def countBracketReversals(inputString):
    length = len(inputString)

    # If the length of the expression is odd, it can't be balanced
    if length % 2 != 0:
        return -1

    stack = []
    for char in inputString:
        if char == '{':
            stack.append(char)
        else:
            # If '}' is encountered and there's an opening '{' in the stack, pop it
            if stack and stack[-1] == '{':
                stack.pop()
            else:
                stack.append(char)  # Otherwise, push '}'

    # At this point, the stack contains mismatched brackets: '}}' or '{{'
    count = 0
    while stack:
        char1 = stack.pop()
        char2 = stack.pop()
        if char1 == char2:
            count += 1  # Count how many pairs are mismatched
        else:
            count += 2  # Two reversals needed to balance one pair

    return count



#main
print(countBracketReversals(stdin.readline().strip()))
```

```python
import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None



def findPathBST(root, data):
    if root is None:
        return None

    path = []

    while root is not None:
        path.append(root.data)

        if data < root.data:
            root = root.left
        elif data > root.data:
            root = root.right
        else:
            break  # Node with data 'k' found

    if root is None:
        return None  # Node with data 'k' not found

    return path[::-1]

def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
data = int(input())
path = findPathBST(root,data)
if path is not None:
    for ele in path:
        print(ele,end=' ')
```

```python
from sys import stdin, setrecursionlimit
setrecursionlimit(10**6)

class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def isIdenticalHelper(node1, node2):
    if not node1 and not node2:
        return True

    if (node1 and not node2) or (not node1 and node2):
        return False

    if node1.data != node2.data:
        return False

    if len(node1.children) != len(node2.children):
        return False

    for child1, child2 in zip(node1.children, node2.children):
        if not isIdenticalHelper(child1, child2):
            return False

    return True

def isIdentical(tree1, tree2):
    return isIdenticalHelper(tree1, tree2)

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i < size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0, childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr1 = list(int(x) for x in stdin.readline().strip().split())
tree1 = createLevelWiseTree(arr1)
arr2 = list(int(x) for x in stdin.readline().strip().split())
tree2 = createLevelWiseTree(arr2)

if isIdentical(tree1, tree2):
    print('true')
else:
    print('false')
```

```python
from math import *
from collections import *
from sys import *
from os import *

import sys

def lengthOfLongestConsecutiveSequence(arr, n):
    visited_elements = {}
    element_to_index_mapping = {}

    for i in range(n):
        element_to_index_mapping[arr[i]] = i
        if arr[i] not in visited_elements:
            visited_elements[arr[i]] = True

    max_length = 0

    for num in arr:
        if num - 1 not in visited_elements:
            current_num = num
            current_length = 0

            while current_num in visited_elements:
                current_length += 1
                current_num += 1

            max_length = max(max_length, current_length)

    return max_length

if __name__ == "__main__":
    # Taking input from command-line arguments
    t = int(sys.argv[1])
    for _ in range(t):
        n = int(sys.argv[2])
        arr = list(map(int, sys.argv[3:]))
        result = lengthOfLongestConsecutiveSequence(arr, n)
        print(result)
```

```python
def subsetSum(l):
    sum_dict = {}  # Dictionary to store the cumulative sum and its corresponding index
    max_length = 0
    current_sum = 0

    for i, num in enumerate(l):
        current_sum += num

        if current_sum == 0:
            max_length = i + 1  # If the sum is zero, the length is the current index + 1

        if current_sum in sum_dict:
            max_length = max(max_length, i - sum_dict[current_sum])
        else:
            sum_dict[current_sum] = i

    return max_length

# Main
n = int(input())
l = list(map(int, input().strip().split()))
print(subsetSum(l))
```
